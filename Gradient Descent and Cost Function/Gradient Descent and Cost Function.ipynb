```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
plt.scatter(x,y)
```

```python
def gradient_descent(x,y):
    m_curr = 0
    b_curr = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.001
    for i in range(iterations):
        y_p = m_curr * x + b_curr
        md = -(2/n) * np.sum(x * (y-y_p))
        bd = (-2/n)* np.sum(y - y_p)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        error = (1/n) * np.sum((y - y_p)**2)
        if i % 100 == 0:
            print(f" m : {m_curr} and b : {b_curr} and Error : {error}")
            plt.plot(x, y_p)
    return m_curr, b_curr
m_curr, b_curr = gradient_descent(x,y)
```

```python
plt.scatter(x, y, color='red')  # Data points
y_final = m_curr * x + b_curr
plt.plot(x, y_final, color='blue')  # Final regression line
plt.xlabel('x')
plt.ylabel('y')
plt.title('Final Linear Regression Line')
plt.show()
```
